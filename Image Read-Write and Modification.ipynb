```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ImageShow(ImageTitle,image):
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(ImageTitle)
    plt.grid()
    plt.show()
```

```python
imageData = {"Image":"image/CatImage.jpg"}
```

```python
img = cv2.imread(imageData["Image"])
```

```python
(height,width,color_ch) = img.shape[:3]
```

```python
print(f"Image Height = {height}\nImage Width = {width}\nImage Color Channel = {color_ch}")
```

```python
ImageShow("Cat Image" , img)
```

```python
# HW---1 : For loop for 25 image with (5 X 5) for show

for i in range(25):
    
    ax = plt.subplot(5,5,i+1)
    ImageShow(f"Cat Image {i}",img)
    
```

```python
# HW---2 : For loop for 25 image with (5 X 5) for save

for i in range(25):
    ax= plt.subplot(5,5,i+1)
    
    cv2.imwrite(f"SaveImage/Cat Image {i}.jpg",img)
    
```

```python
image = cv2.imread(imageData["Image"])
image
```

```python
#HW----4

red = (0,0,255)
green = (0,255,0)
blue = (255,0,0)

cv2.rectangle(image,pt1 = (180,150) , pt2 = (150,175) ,color=red,thickness = 3)
cv2.putText(image,org=(170,150),fontScale=0.4,color = red,thickness = 1, text = "Right Eye",fontFace = cv2.FONT_ITALIC)

cv2.circle(image,(260,160),30,green,2)
cv2.line(image,pt1=(240,240),pt2 = (180,240) , thickness = 6 ,color=blue)
ImageShow("Drow Rectangle,Circle,Line",image)
```

```python
movementInfo = np.float32([[1,0,-125],[0,1,-135]])
updatedImageAfterMovement = cv2.warpAffine(image,movementInfo,(image.shape[1],image.shape[0]))
```

```python
ImageShow("Updated Image After Movement",updatedImageAfterMovement)
```
